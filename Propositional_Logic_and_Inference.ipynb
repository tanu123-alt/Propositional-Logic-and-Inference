```python
from sympy import symbols
from sympy.logic.boolalg import Implies, And, Not, Or
from sympy.logic.inference import satisfiable

# 1. Define the propositional symbols
x, y, z = symbols('x y z')

# 2. Construct a propositional logic expression
# Example: (x => y) & (y | z) & (~z)
# Using the classes:
expr_classes = And(Implies(x, y), Or(y, z), Not(z))

# Using Python operators (requires careful use of parentheses):
expr_operators = (x >> y) & (y | z) & (~z)

# 3. Check satisfiability
model_classes = satisfiable(expr_classes)
model_operators = satisfiable(expr_operators)

print(f"Expression with classes is satisfiable: {model_classes}")
print(f"Expression with operators is satisfiable: {model_operators}")

# Example of an unsatisfiable expression (a contradiction)
unsat_expr = And(x, Not(x))
model_unsat = satisfiable(unsat_expr)
print(f"Unsatisfiable expression is satisfiable: {model_unsat}")

# Example using all_models=True to get a generator of all models
models_generator = satisfiable((x >> y) & y, all_models=True)
print("Models for (x => y) & y:")
for model in models_generator:
    if model:
        print(f"  Model: {model}")
    else:
        print("  Unsatisfiable (this should not appear here)")

# To use the algorithm parameter (e.g., with "z3" if installed), you can specify it
# Note: "z3" might require the z3-solver library to be installed separately
# satisfiable(expr_operators, algorithm="z3")
```
